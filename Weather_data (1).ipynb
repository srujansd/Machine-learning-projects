```python
#Calculating the amount of data or say no of rows in the dataset before removing the rows containing null values
before_rows=data.shape[0]
print(before_rows)
```

```python
# delete the null values

data = data.dropna()
```

```python
#Calculating the amount of data or say no of rows in the dataset after removing the rows containing null values
after_rows = data.shape[0]
print(after_rows)
```

```python
before_rows - after_rows
```

```python
#filter values which contains more than 24.99 relative humidity at 3pm
clean_data=data.copy()
clean_data['high_humidity_label'] = (clean_data['relative_humidity_3pm'] > 24.99) *1
print(clean_data['high_humidity_label'])
```

```python
y=clean_data[["high_humidity_label"]].copy()
y.head()
```

```python
clean_data['relative_humidity_3pm'].head()
```

```python
y.head()
```

```python
#storing all the morning features other than humidity at 3pm in the morning feature
morning_features = ['air_pressure_9am','air_temp_9am','avg_wind_direction_9am','avg_wind_speed_9am',
        'max_wind_direction_9am','max_wind_speed_9am','rain_accumulation_9am',
        'rain_duration_9am']
```

```python
#copying the values from the clean_data dataset to new dataset "x" which is consist only morning feature data
x=clean_data[morning_features].copy()
x.columns
```

```python
y.columns
```

```python
#perform test and train and split

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=324)
```

```python
from sklearn.tree import DecisionTreeClassifier

humidity_clssifier = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
humidity_classifier.fit(X_train, y_train)
```

```python
type(humidity_classifier)
```

```python
#testing the model
y_predicted = humidity_classifier.predict(X_test)
```

```python
y_predicted[:10]
```

```python
accuracy_score(y_test,y_predicted)*100
```
